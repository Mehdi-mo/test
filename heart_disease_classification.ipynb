```python
# we want to get the patients over 50

over_50 = pd.Series
over_50
```

```python
#scatter plot of age and cholestrol
over_50.plot.hist(x='chol', y='age')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
over_50.plot(kind='scatter',
            x='chol', 
            y='age',
            c='target',
            ax=ax)
ax.set_ylim(45, 100)
```

```python
fig ,ax = plt.subplots(figsize=(5,5))
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'],
                    c=over_50['target'])
ax.set_xlim(45, 100)
ax.set(title='comparing age and cholestrol level', xlabel='age', ylabel='chol')

# add legend
ax.legend(*scatter.legend_elements(), title='target')

ax.axhline(over_50['chol'].mean(), color='red', linestyle='--')
```

```python
# subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2 ,ncols=1, figsize=(5, 10))


fig.suptitle('the heart disease analysis', fontsize=16, fontweight='bold')

ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
ax0.set(title='comparing age and cholestrol level', xlabel='age', ylabel='chol')
ax0.axhline(y=over_50['chol'].mean(), color='red', linestyle='--')
ax0.legend(*scatter.legend_elements(), title='target')

ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])
ax1.set(title='comapring age and thalach', xlabel='age', ylabel='thalach')
ax1.axhline(y=over_50['thalach'].mean(), color='red', linestyle='--')
ax1.legend(*scatter.legend_elements(), title='target')
```
